```
#open the text file:
text_file = open("women empowerment.txt")

#read the data
text = text_file.read()

#datatype of data read:
print(type(text))
print("\n")

#print the text:
print(text)
print("\n")

#length of the text:
print(len(text))
```

```
import nltk
nltk.download('all')
from nltk import word_tokenize
from nltk import sent_tokenize
```

sentence tokenization


```
sentences = sent_tokenize(text)
sentences
type(sentences)
```

word tokenization


```
word = word_tokenize(text)
word
len(word)
```

find the frequency distribution

```
from nltk.probability import FreqDist
```

```
fdist = FreqDist(word)
fdist.most_common(10)
```

plot the frequency of the words before removing stopwords


```
import matplotlib.pyplot as plt
```

```
fdist.plot(10)
```

removing punctuation marks

```
not_punc_word = []
for w in word:
  if(w.isalpha()):
    not_punc_word.append(w.lower())
print(not_punc_word)
print(len(not_punc_word))
```

plotting graph without punctuation marks

```
fdist = FreqDist(not_punc_word)
fdist.most_common(10)
fdist.plot(10)
```

removing stopwords

```
from nltk.corpus import stopwords

stopwords = stopwords.words("english")
print(stopwords)
```

```
clean_words = []
for w in not_punc_word:
  if(w not in stopwords):
    clean_words.append(w)
print(clean_words)
len(clean_words)
```

removing extra words after stopwords

```
x = ['also', 'still', 'must']
final_words = []
for a in clean_words:
  if a not in x:
    final_words.append(a)
print(final_words)
len(final_words)
```

final frequency distributer

```
freq = FreqDist(final_words)
freq.most_common(10)
freq.plot(10)
```

**word cloud implementation**

```
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image

women_mask = np.array(Image.open("Screenshot (107).png"))
#Generating the wordcloud:
wordcloud = WordCloud(background_color="white", mask=women_mask,
                      stopwords=stopwords, contour_width=0.5, contour_color='black'
                      ,max_font_size=40, relative_scaling=0).generate(text)

#plot the wordcloud:
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud)
plt.title("Women Empowerment")

#to remove the axis value:
plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
```
